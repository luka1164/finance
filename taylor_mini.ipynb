```python
database = pd.DataFrame(index=main_df.index)
database = main_df.copy()

for ticker in tickers:
    database['5sma-{}'.format(ticker)] = database['{}'.format(ticker)].rolling(5).mean()
    
database.dropna(inplace=True)    
database.head()
```

```python
import numpy as np

day_database = pd.DataFrame()

for ticker in tickers:
    database['days_sma{}'.format(ticker)] =  np.where(database['{}'.format(ticker)]>database['5sma-{}'.format(ticker)], 1, -1)
    day_database['days_sma{}'.format(ticker)] = database['days_sma{}'.format(ticker)]
    
bias = day_database.iloc[-10:].sum()
bias
```
